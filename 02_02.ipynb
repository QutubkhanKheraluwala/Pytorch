```python
import torch
```

```python
first_tens = torch.tensor([[13,2,5,31],[12,7,6,4]])
second_tens = torch.tensor([[11,1,4,32] , [10,5,3,4]])
```

```python
add_tens = first_tens + second_tens

print(add_tens)
print(add_tens.size())
```

```python
sub_tens = first_tens - second_tens
print(sub_tens)
print(sub_tens.size())
```
