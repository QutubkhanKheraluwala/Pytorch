```python
if torch.cuda.is_available():device = "cuda"
else:device = "cpu"
print(device)
```

```python
tens_a = torch.tensor([[12,13,21,6] , [10,2,4,1]] , device=device)
tens_b = torch.tensor([[4,5,9,12] , [11,7,5,9]] , device=device)

mul_tens = tens_a * tens_b
print(mul_tens)
```
